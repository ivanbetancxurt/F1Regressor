```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
rf_results = pd.read_csv('data/results/randomforest_max_depth-23_n_estimators-700.csv')
errors = rf_results['predicted'] - rf_results['actual']

plt.figure()
plt.scatter(rf_results['actual'], rf_results['predicted'], alpha=0.6)
plt.plot(
    [rf_results['actual'].min(), rf_results['actual'].max()],
    [rf_results['actual'].min(), rf_results['actual'].max()],
    linestyle='--',
    color='red'
)
plt.xlabel('Actual Finishing Position')
plt.ylabel('Predicted Finishing Position')
plt.title('Actual vs Predicted [Random Forest Regressor]')
plt.show()
```

```python
plt.figure()
hb = plt.hexbin(rf_results['actual'], rf_results['predicted'], gridsize=30)
plt.colorbar(label='Count')
plt.plot(
    [rf_results['actual'].min(), rf_results['actual'].max()],
    [rf_results['actual'].min(), rf_results['actual'].max()],
    linestyle='--',
    color='red'
)
plt.xlabel('Actual Finishing Position')
plt.ylabel('Predicted Finishing Position')
plt.title('Density of Actual vs. Predicted [Random Forest Regression]')
plt.show()
```
